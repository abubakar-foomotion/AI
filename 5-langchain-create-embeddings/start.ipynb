# Embeddings via Ollama
### Create embeddings using nomic-embed-text model
#### we are just creating embeddings of a string here

```python
from langchain_ollama import OllamaEmbeddings

# Initialize embedding model
embeddings = OllamaEmbeddings(model="nomic-embed-text")

# Create embedding
vector = embeddings.embed_query("Hello world")

# print(vector[:5])  # just first 5 numbers
print(len(vector))
```

#### we create embeddings of the entire documents

```python
texts = [
    "LangChain is powerful",
    "Ollama runs models locally",
    "Embeddings convert text to vectors"
]

vectors = embeddings.embed_documents(texts)

print(len(vectors))        # 3 vectors
print(len(vectors[0]))     # vector size
```

# Embeddings via Hugging Face

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)

vector = embeddings.embed_query("Hello world")

print(vector[:5])
```
