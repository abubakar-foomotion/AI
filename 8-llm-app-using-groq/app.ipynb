# LLM application for langugage translation
here actually we are using groq for models for translation whose api key should be in the .env file but we are not currently having api key as groq.com is not responding

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os

load_dotenv()

llm = ChatGroq(
    model="llama3-8b-8192",  # or mixtral
    temperature=0
)
```

```python
from langchain.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a professional translator."),
    ("human", "Translate the following text into {target_language}: {text}")
])

# NOTE: Each tuple inside the the list to the ChatPromptTemplate is : (role, content)
# and these roles are good to have system for the model to behave like and human to tell the model
# this is human given input
```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()
```

```python
chain = prompt | llm | parser
```

```python
response = chain.invoke({
    "target_language": "Urdu",
    "text": "How are you?"
})

print(response.content)
```
