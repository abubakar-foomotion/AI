```python
#DICTRIONARY
data = {
    "name" :"abubakar",
    "age"  :25,
    "company" : "foomotion",
}
print(data["name"])
for key,value in data.items():
    print(f"{key} => {value}")

#NESTED DICTIONARIES
apiData = {
    "p1" : {"name" : "a" , "age":1},
    "p2" : {"name" : "b" , "age":2},
    "p3" : {"name" : "c" , "age":3},
}
for key,value in apiData.items():
    print(f'{key} =--= ')
    for nestedKey,nestedValue in value.items():
        print(f'{nestedKey} => {nestedValue}')
```
