| Splitter                       | How it splits                                  | Intelligence Level | Best For                | Risk                  |
| ------------------------------ | ---------------------------------------------- | ------------------ | ----------------------- | --------------------- |
| CharacterTextSplitter          | Fixed chars + separator                        | ❌ Low              | Simple text             | Breaks meaning        |
| RecursiveCharacterTextSplitter | Tries multiple separators (para → line → word) | ✅ High             | General use (PDF, text) | Slightly slower       |
| HTMLHeaderTextSplitter         | Based on HTML headings                         | ✅ Very High        | Web pages, docs         | Depends on clean HTML |
| RecursiveJsonSplitter          | Based on JSON structure                        | ✅ Very High        | APIs, structured data   | Needs valid JSON      |





🧠 1) High-Level Comparison + Summary
🔍 Big Picture

All splitters do one job:

👉 Convert large data → meaningful smaller chunks

But the difference = HOW they decide boundaries

# Splitting via RecursiveCharacterTextSplitter
it is intellegent in spilliting

```python
# STEP 1: => LOADING THE PDF OR DOUCMENT IN THE DOCUMENT FORMAT
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("climate.pdf")
docs = loader.load()


print(docs[0].page_content)
print(docs[0].metadata)
```

```python
print(type(docs))
print(type(docs[0]))
```

```python
# STEP 2: => SPLITTING
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Split
splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=150
)

chunks = splitter.split_documents(docs)

print(f"Original docs: {len(docs)}")
print(f"Chunks created: {len(chunks)}")
```

```python
print("-------------------------------------------------------------------------------------")
print(chunks[0])
print("-------------------------------------------------------------------------------------")
print(chunks[1])
print("-------------------------------------------------------------------------------------")
```

# CharacterTextSplitter
👉 It splits text based purely on:

characters
a separator (like \n, space, etc.)

❗ It does NOT care about meaning
⚠️ Limitations (VERY IMPORTANT)
❌ Breaks sentences randomly
❌ Context loss
❌ Not ideal for LLM apps

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter

loader = TextLoader("sample.txt")
docs = loader.load()

splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=100,
    chunk_overlap=20
)

chunks = splitter.split_documents(docs)

for chunk in chunks:
    print(chunk.page_content)
    print("------")
```

# HTMLHeaderTextSplitter
splits HTML based on headings

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<h1>AI</h1>
<p>Artificial Intelligence is growing.</p>

<h2>Applications</h2>
<p>Used in healthcare and finance.</p>
"""

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
]

splitter = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)

chunks = splitter.split_text(html_string)
url_chunks = splitter.split_text_from_url("example.com")

for chunk in chunks:
    print(chunk.page_content)
    print(chunk.metadata)
    print("------")
```

# RecursiveJsonSplitter

```python
from langchain_text_splitters import RecursiveJsonSplitter

json_data = {
    "users": [
        {
            "name": "Ali",
            "skills": ["Python", "AI", "ML"]
        },
        {
            "name": "Ahmed",
            "skills": ["JavaScript", "React"]
        }
    ]
}

splitter = RecursiveJsonSplitter(max_chunk_size=5)

chunks = splitter.split_json(json_data)
print(chunks)

# for chunk in chunks:
#     print(chunk)
#     print("------")
```
